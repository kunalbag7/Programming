```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('lenna.jpg',0)
plt.imshow(img,'gray')
print(img)
```

```python
img.shape
```

```python
def zero_or_one(img, center, i, j): 
      
    val1=0
      
    try:  
        if img[i][j] >= center: 
            val1 = 1
              
    except:  
        pass
      
    return val1
```

```python
lbp_img = np.zeros((img.shape[0],img.shape[1]), np.uint8)

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        center=img[i][j]
        val=[]
        
        val.append(zero_or_one(img, center, i-1, j-1)) 
        val.append(zero_or_one(img, center, i-1, j))  
        val.append(zero_or_one(img, center, i-1, j+1))  
        val.append(zero_or_one(img, center, i, j+1))  
        val.append(zero_or_one(img, center, i+1, j+1))  
        val.append(zero_or_one(img, center, i+1, j))  
        val.append(zero_or_one(img, center, i+1,j-1))  
        val.append(zero_or_one(img, center, i, j-1))
          
            

            
        v=0
        for k in range(len(val)):
            v += val[k] * pow(2,(7-k))
        
        
        lbp_img[i][j]=v
        
            
                
            
```

```python
print(lbp_img)
plt.imshow(lbp_img,"gray")
#plt.show()
```

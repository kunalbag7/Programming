```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
img = cv2.imread('sample3.jpg',0)
# img=cv2.resize(img,(512,512))
# fig=plt.figure(figsize=(15,15))
# ax=fig.add_subplot(111)
plt.imshow(img,'gray')
print(img)
```

```python
(thresh, blackAndWhiteImage) = cv2.threshold(img, 49, 255, cv2.THRESH_BINARY)
plt.imshow(blackAndWhiteImage,'gray')
```

```python
padded_arr =np.zeros((img.shape[0]+2,img.shape[1]+2))
for i in range (img.shape[0]):
    for j in range(img.shape[1]):
        padded_arr[i+1][j+1] = blackAndWhiteImage[i][j]

struc_ele=np.array([[255,255,255],[255,255,255],[255,255,255]])
new_img=np.zeros(img.shape,'uint16')
for i in range (img.shape[0]):
    for j in range(img.shape[1]):
        if (np.alltrue(np.equal(struc_ele,padded_arr[i:i+3,j:j+3]))):
            new_img[i][j]=255
        else:
            new_img[i][j]=0
```

```python
plt.figure(figsize=(15,15))
plt.subplot(1,2,1)
plt.imshow(img,'gray')
plt.title('Real Image')
plt.subplot(1,2,2)
plt.imshow(new_img,'gray')
plt.title('After Erosion')
```

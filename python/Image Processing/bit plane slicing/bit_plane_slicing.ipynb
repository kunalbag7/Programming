```python
start=time.time()
bin = []
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
         bin.append(np.binary_repr(img[i][j] ,width=8))
            
for m in range (8):
    bit_img=(np.array([int(i[m]) for i in bin],dtype = np.uint8) * pow(2,(7-m)) ).reshape(img.shape[0],img.shape[1])
    plt.imshow(bit_img,'gray')
    plt.show()
    
finish=time.time()
print(finish-start)
```

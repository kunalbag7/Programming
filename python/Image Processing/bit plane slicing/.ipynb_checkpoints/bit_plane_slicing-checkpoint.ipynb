```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('Lenna.png',0)
plt.imshow(img,'gray')
```

```python
img3=cv2.imread('Lenna.png')
img2 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
img.shape
```

```python
iar = np.asarray(img)
print(iar)
```

```python
bin = []
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
         bin.append(np.binary_repr(img[i][j] ,width=8))
```

```python
eight_bit_img = (np.array([int(i[0]) for i in bin],dtype = np.uint8) * 128).reshape(img.shape[0],img.shape[1])
plt.imshow(eight_bit_img,'gray')
```

```python
seven_bit_img = (np.array([int(i[1]) for i in bin],dtype = np.uint8) * 64).reshape(img.shape[0],img.shape[1])
plt.imshow(seven_bit_img,'gray')
```

```python
six_bit_img = (np.array([int(i[2]) for i in bin],dtype = np.uint8) * 32).reshape(img.shape[0],img.shape[1])
plt.imshow(six_bit_img,'gray')
```

```python
five_bit_img = (np.array([int(i[3]) for i in bin],dtype = np.uint8) * 16).reshape(img.shape[0],img.shape[1])
plt.imshow(five_bit_img,'gray')
```

```python
four_bit_img = (np.array([int(i[4]) for i in bin],dtype = np.uint8) * 8).reshape(img.shape[0],img.shape[1])
plt.imshow(four_bit_img,'gray')
```

```python
three_bit_img = (np.array([int(i[5]) for i in bin],dtype = np.uint8) * 4).reshape(img.shape[0],img.shape[1])
plt.imshow(three_bit_img,'gray')
```

```python
two_bit_img = (np.array([int(i[6]) for i in bin],dtype = np.uint8) * 2).reshape(img.shape[0],img.shape[1])
plt.imshow(two_bit_img,'gray')
```

```python
one_bit_img = (np.array([int(i[7]) for i in bin],dtype = np.uint8) * 1).reshape(img.shape[0],img.shape[1])
plt.imshow(one_bit_img,'gray')
```
